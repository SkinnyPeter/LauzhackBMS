```python
import pandas as pd
import os

files = ["BRISTOR_Zegoland", "INNOVIX_Elbonie", "INNOVIX_Floresland"]

for file_name in files:
    # Load the Excel file
    input_file = os.path.join("files", f"{file_name}.xlsx")  # Replace with your .xlsx file path
    if not os.path.isdir(os.path.join("files", f"{file_name}_as_csv")):
        os.mkdir(os.path.join("files", f"{file_name}_as_csv"))
    
    # Load the Excel file and inspect the sheet names
    excel_data = pd.ExcelFile(input_file)
    sheet_names = excel_data.sheet_names  # List of sheet names
    
    print("Found sheets:", sheet_names)
    
    # Convert each sheet to a separate CSV
    for sheet_name in sheet_names:
        df = pd.read_excel(input_file, sheet_name=sheet_name)
        output_file = f"{sheet_name}.csv"  # Save with sheet name as file name
        df.to_csv(os.path.join("files", f"{file_name}_as_csv", output_file), index=False)
        print(f"Saved sheet '{sheet_name}' to {output_file}")
```
